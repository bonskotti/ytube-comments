## 2. Preprocessing

```python
# check for duplicates
data['Comment Id'].value_counts()[:10]
```

```python
# remove duplicates
data.drop_duplicates(['Comment Id'], keep='first', inplace=True)
```

```python
# check result
data['Comment Id'].value_counts()[:10]
```

### Standardization

```python
like_score = preprocessing.StandardScaler().fit_transform(data[['Likes']])
```

```python
print(max(like_score), min(like_score))
print(like_score)
```

```python
y_not_zero = data['Likes'].apply(lambda x: x + 1e-10)
```

```python
like_score_lg = np.log(y_not_zero)
```

```python
print(like_score_lg)
```

```python
print(max(like_score_lg), min(like_score_lg))
```
